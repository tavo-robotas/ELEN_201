# Medžiagų savybės ir Elektrinės grandinės principas



Skirtingų medžiagų elektronai turi skirtingą judėjimo laisvę. Tai reiškia kad medžiagose kaip Stiklas ar Guma Valentiniai elektronai skriejantys aplinkui branduolį negalės atsiplėšti nuo savo branduolio. Kitose medžiagose kaip Sidabras, Elektronai gali laisvai keliauti tarp atomų. Dėl tokio elektronų judėjimo jie yra skaitomi **laisvaisiais elektronais** (ang. "free electrons"). Žemiau pavaizduotame pavizdyje pavaizduota laisvūjų elektronų judėjimo principas 

Šis laisvūjų elektronų judėjimas nurodo medžiagų elektros pralaidumo savybes. Pačios medžiagos yra skaidomos į tris pagrindies grupes pagal savo laidumą elektrai:
- Laidininkai - Metalai
- Puslaidininkiai - Pusmetaliai
- Izoliatoriai - Ne metalai



### Medžiagų savybės

**Laidininkas**
- Perijodinėje lentelėje visi esantys metalai yra laidininkai. 
- Medžiaga kuri leidžia elektronams lengvai tekėti pro save 
- mažas pasipriešinimas elektronų tekėjimui
- Naudoja Laidžių laidų ir kabelių gamyboje

**izoliatoriai**
- Perijodinėje lendelėje pažymėti kaip nemetalais (ang. NonMetals)
- Medžiagos kurios neleidžia elektronams tekėti pro save
- Labai aukštas pasipriešinimas
- Naudojamas elektros ir elektroninių prietaisų izoliacijai, elektros šoko prevencijai.

**Puslaidininkis**
- Perijodinėje lentelėje jie žymimi kaip pusmetaliais
- Medžiaga kurios laidumo savybės yra tarp laidininko ir izoliatoriaus 
- Vidutinis prasipriešinimas elektronų tėkmei
- Naudojami gaminant Elektronikos komponentus

### Elektronų tėkmė

Kaip ir buvo minėta prieš tai kad elektronų tėkmė laidininke yra laisva. Tai reiškia kad jie keliauja laisvai pro atomų valentinius sluoksnisu be jokio specifinio greičio ar krypties. Bet elektronai gali būti paveikti kad jie judėtų per laidžią medžiagą į mums norimą kryptį. Šitas Kryptingas elektronų judėjimas yra vadinamas **elektra** arba **elektros srovė**.

Elektronai gali laisvai judėti pro atomų struktūroje esančius tarpus. Visos medžiagos gal ir atrodo solidžios bet ištikrūjų jos yra pilnos tuščios erdvės

Verta paminėti kad visi esantys elektronai juda tuo pačiu momentu, vienas elektronas stumia priekyje esantį elektrona o tas pajudintas elektronas stumia kitą priekyje esantį elektroną. Taip visi šie elektronai juda vienu metu. 
Elektronų srautas per visą laidininką judą į priekį ir sustoja beveik akimirksniu. Panaši analogija būtų su kamuoliukais vamzdyje.


### Elektronų tėkmė laide

Elektros srovė gali tekėti tik ten kur yra tiesioginis kelias elektronams tekėti. 
Kamuoliukų pavizdyje, Kamuoliukai gali judėti tik į vieną vamzdžio pusę ir galiausiai pro vamzdį tik tuo atveju jei kitas vamzdžio galas yra atviras ir tik tokiu būdu kamuoliukai gali išlysti kitame vamzdžio gale. Jeigu kitas vamzdžio galas bus uždarytas tai tada kamuoliukai susigrūs visame vamzdyje ir jie nebeturės kur keliauti, Kamuoliukų srovė nutrūks. Tas pats galioja ir elektrinėje srovėje, pastovi elektronų srovė reikalauja salygos kad kelias būtų nutrūkęs ir pastovus.


Plona linija yra tokia kuri yra pavaizduota aukščiau yra sutartinis ženklas kuris nurodo vientisą induktyvų laidą. Kadangi laidas yra pagamintas iš induktyvos medžiagos, tokios kaip varis, jį sudarantys atomai turi daugybe laisvųjų elektronų kurie gali laisvai keliauti laidu. Norint kad pro šį laidą tekėtu elektronų srovė prie jo vieno galo reikia prijungti elektronų šaltinį o prie kito galo vietą pro kur tiems elektronams pasišalinti. Dėl paprastumo šiuo metu šio laido gallus pavadinkime “Elektronų Šaltiniu” ir “Elektronų išeiga”. 

Rodyklės žymi kuria kryptimi teka elektronai laide iš šaltinio į Elektros išeigą. Norint kad elekronai tekėtų jiems reikia tiesioginio kelio judėti pirmyn. Tai jeigu nutrūks šis mūsu laidas tai elektronams nebus kur keliauti ir todėl elektros srovė nevyks


Tarp dviejų nutrūkusių medžiagų šiuo metu yra oro tarpas kuris neleižia elektronams tekėti. Šis tarpas nutraukia elektronų srautą iš šaltinio į Elektronų išeigą. Kalbant apie elektrą, kai laidas buvo vientisas, mes turėjome nenutrūkstamą elektronų kelią. Nupjaunant ir atskyrus laidą, šis tęstinumas nutraukiamas.

Jei prijungsime kitą laido gabalą nuo šaltinio iki paskirties vietos, sukurdami fizinį kontaktą (metalas su metalu) tarp vielos dalių, atkursime nenutrūkstamą elektronų tekėjimo kelią. Du taškai parodyti žemiau žymi šį fizinį laidų kontaktą.

### Kas yra elektros grandinė?

Kad ši Elektronų šaltinio ir paskirties išeigos idėja veiktų, abiem galams reikės begalinės talpos, pavyzdžiui, naudojant kamuoliukų ir vamždžio analogijai reiktų be galo didelių kibirų nuolatiniam rutuliukų srautui. Norint išvenkti šios begalinės talpos koncepto mes galime sujungti vieno laido ar vamzdžio galus su kitu. Taip visi mūsu "elektronai" judės ratu

Kiekvienas Elektronas nešantis krūvį, judantis pagal laikrodžio rodyklę šioje grandinėje, stumia priešais esantį, o šis stumia priešais esantį ir t.t. Dabar mes galime palaikyti nuolatinį elektronų srautą neribotą laiką, nereikalaujantį begalinių talpų. Viskas ko mums reikia šiam srautui palaikyti, yra nuolatinė šių elektronų motyvavimo priemonės, kurią aptarsime vėliau.

### Ką reiškia kai elektronų grandinė yra nutrūkusi?

Tęstinumas grandinėje yra toks pat svarbus kaip ir prieš tai minėtame pavyzdyje su tiesia viela, jungiančia **Elektronų šaltinį** ir **Elektronų išeiga**, bet koks grandinės nutraukimas, nepaisant vietos, trukdo elektros krūviui tekėti. Šis principas pabrėžia, kad be nenutrūkstamos laidžios medžiagos uždaro rato negalima išlaikyti nuolatinio elektronų srauto.

# Įtampa ir Srovė

Norint geriau ir aiškiau suprasti kas yra **Srove (Elektronų tėkmė)** ir **Įtampa** mes galime tai gerai vizualizuoti vandens bokšto pavyzdžiu.


Vandens bokšte esantis vanduo buvo pripildytas vandens siurbliu iš netoliese esančio tvenkinio. Bokšte esantis vanduo yra veikiamas gravitacijos kuri sukuria energiją. Ši energija bando ištrūkti ir sugrįžti į savo pradine padetį tvenkinyje. Jeigu mes nutiesime atitinkamą vamzdį iš bokšto į tvenkinį, gravitacijos veikiamas vanduo bus stumiamas atgal į tvenkinį.

Norint pakelti siurbliu vandenį aukštyn iš žemos padėties į aukštesnę yra reikalinga Energija. Ši energija yra laikoma bokšte tol kol iš jo neteka vanduo atgal į ežerą. Paleidus vandenį žemyn, užpumpuota energija yra paleidžiama ir visa užlaikyta energija dabar grįžta atgal į ežerą. 

Jeigu bokštas būtų aukštesnis tai pareikalautų iš pompos dar daugiau energijos norint pakelti vandenį. o išleistas vanduo gryžtantis į ežerą turės dar daugiau energijos

Elektronai veikia panašiu principu. Kai įkrauname objektus neigiamu krūviu mes "užpumpuojame" jį elektronais. Šie elektronai dabar nori sugrįžti į savo pradinę padėtį **sudarydami balansą atomuose**. <br> Jėga traukianti elektronus į jiems skirtas pozicijas aplinkui atomo branduolius yra panaši į gravitacijos jėga. Ši jėga veikia vandenį bokšte ir bando vandenį nuleisti į jo pradinę padėtį ežere.

- Lygiai taip pat kaip užpumpuotas vanduo bokšte lemia sukauptą energija. Taip "užpumpuoti" elektronai sukuria elektros krūvio disbalansą, šiame disbalanse sukaupiamas tam tikras energijos kiekis.
- Ir lygiai taip pat suteikiant galimybę vandeniui tekėti atgal iš bokšto aukštumų, išleidžiama ta sukaupta energija, taip elektronams suteikiant galimybę tekėti atgal į pradinį „lygį“, išleidžiama sukaupta energija. 
- Kai elektronai yra laikomi tokioje būsenoje (kaip ir vanduo laikomas aukštai bokšte), bokšte laikoma sukaupta energija yra vadinama **potencialia energija**. Mes taip ją vadiname nes energija nėra dar iki galo išsilaisvinusi ir turi potencialą išsilaisvinti.

### Įtampos konceptas

Kai krūvio nešėjai (elektronai) yra tokioje statinėje būsenoje (kaip vanduo esantis aukštai rezervuare), ten sukaupta energija vadinama potencialia energija, nes turi galimybę (potencialą) išsilaisvinti, kuri dar nėra iki galo realizuota.<br> Kai vaikštai ant kilimo sausą dieną basomis kojomis tu sukuri krūvio disbalansą tarp saves ir kilimo. Pati trintis kurią sukeli eidamas kilimu ir sukelia krūvių disbalansą ir elektronai nuo kilimo susikaupia ant odos. Šis krūvis vadinamas statine elektra kurios žmogus net nepajus kad ši energija susikaupė ant jo. Tačiau jeigu palies metaline durų rankena visas krūvis kuris yra laikomas ant odos dabar bus greitai paleistas pro ranką ir pasijus elektros srovę.  

Ši potenciali energija laikoma elektrinio krūvio formoje gali būti išreikšta terminu **Voltas (V).** Voltas techniškai yra potencialios energijos vienetas

### Įtampos apibrėžimas

Elektros energijos šaltiniuose įtampa nurodo potencialią energiją, turimą vienam elektros srovės vienetui, kad krūvis (elektros srovė) būtų išstumtos per laidininką. Tai reiškia energijos išsiskyrimo galimybę, kai krūviai juda tarp dviejų taškų. Kadangi įtampa yra potencialios energijos išraiška, nurodanti energijos išsiskyrimo galimybę arba potencialą, kai krūvis juda iš vieno „lygio“ į kitą. Įtampa visada nurodoma tarp dviejų taškų. Atsižvelgiant į vandens bokšto analogija:

Dėl aukščių skirtumo antroje pavaizduotame "lygije" (#2 lygis) susidarys didesnis energijos potencialas negu pirmoje vietoje (#1 lygis). Šitas principas gali būti išreikštas ridenant akmenį nuo kalno: Didesnis akmuo turės stipresnį poveikį aplinkai, jei šis akmuo bus pakankamai didelis tai jis nuo kalno nusineš viską kas pakliuvo jo kelyje, Jei leisime mažam akmeniui redėti kalnu tai jis neturės tiek griaunamosios jėgos. 

Negalime įvertinti vandens bokšte sukauptos energijos kiekio paprasčiausiai išmatuodami vandens tūrį, nei galime numatyti krintančios uolos smūgio stiprumą vien žinant uolos svorį: Abėjuose pavyzdžiuose mes turim atsižvelkti kaip toli jie kris nuo jų pradinės pozicijos. Energijos kiekis, išsiskiriantis leidžiant masei nukristi, yra santykinis su atstumu tarp jos pradžios ir pabaigos taškų. Todėl įtampa visada išreiškiama kaip dydis tarp dviejų taškų. Įdomu tai, kad masės, galinčios „nukristi“ iš vieno aukščio į kitą, analogija (išsireiškimas) yra toks tinkamas modelis, kad įtampa tarp dviejų taškų kartais vadinama **įtampos kritimu.** (Voltage drop)

### Įtampos generavimas

Įtampa gali būti sugeneruojama netik trinant vienus objektus su kitais. Cheminės reakcijos, spinduliuotės energija, ir magnetizmo įtaka laidininkams yra keli būdai kaip mes galime generuoti įtampa. Atitinkami pavyzdžiai yra Baterijos, saulės scelės ir generatoriai. Kas svarbiausia šiuo metu  suprasti tai kad įtampos šaltinius mes galime naudoti elektrinėse grandinėse. Apačioje mes matome įtampos šaltinį - baterija.

įtampos šaltiniuose tokiuose kaip baterijos visada mes matysime vieną kontaktą pažymėta (+) o kitą (-). Šie kontaktai turi savo pavadinimus vienas yra **Anodas** žymėtas **(+)** kitas **Katodas** žymimas **(-)**. <br> Jeigu baterija nesudarys uždaro rato tai tarp Katodo ir Anodo įtampa vistiek bus bet elektros srovė netekės. 

Tas pats principas galioja ir vandens bokšto pavizdyje be gryžtamojo vamzdžio atgal į vandens tvenkinį, užlaikyta energija bokšte negalės išsilaisvinti. Kai vandens bokštas yra visiškai užpildytas, vandens srovė (energija) negali atsirasti, nesvarbu,kaip stipriai siurblys bandys pripildyti bokštą. <br> Tam kad srovė atsirastu Mes galime prijungti laido galą prie baterijos dviejų terminalų taip sudarydami uždarą ratą elektronams judėti. Pačią elektros srovę pažymėsim pagal laikrodžio rodykle iš Anodo į Katodą.

### Elektros srovės konceptas

Tol kol baterija turi įtampos ir prie jos prijungtas laidininkas sudaro uždarą ratą tol elektros srovė tekės pro šį uždarą ratą. **Šis tolygus elektronų judėjimas pro laidą vadinama srove, žymimas (I)**. Tol kol įtampa "stumia" į tą pačia puse tai ir patys elektronai judės į tą pačia puse. toks tolygis vienapusis elektronų judėjimas vadinamas **nuolatine srove (DC)**. Vėliau susipažinsime su **Kintamaja srove (AC)** Elektros srove kuri keičia savo judėjimo krypti.

Kadangi Elektros srovę sudaro atskiri elektronai judandys viena kryptimi per laidininką ir šie elektronai stumia vienas kita kaip ir vanduo tekantis vamzdyje tai srovė visoje grandinėje bus vienoda. Jeigu mes norėtume patikrinti elektrinės grandinėje tekančių elektronų skaičių duotame laiko tarpe tai mes pastebėtume kad betkurioje elektros grandinės vietoje prabėgančių elektronų skaičius yra lygiai toks pats betkurioje grandinės vietoje. Nepaisant induktoriaus ilgio ar storio šis principas vistiek galioja

Jeigu mes nutrauktume elektrinės grandinės testinumą betkurioje grandinės vietoje tai elektros srovės nebebūtu visoje grandinėje. O visa baterijos sukurta įtampa pasireikš nutrūkusiuose galuose:

### Įtampos kritimo (Voltage drop) polerizacija

Apačioje esančiame pavizdyje mes matome **(+)** ir **(-)** ženklus prie nutrūkusios grandinės galų, matome kad šie ženklai atitinka prie baterijos jungčių esančius ženklus. Šitie ženklai nurodo krypti kur įtampa bando stumti elekrtos srovę. Tokia **Įtampos Potencialo** kryptis dažniausiai vadinama **Polerizacija**. 

Prieštai rodytuose pavyzdžiuose elektros grandinės nėra praktiškos. Tiksliau tokios grandinės netgi gali būti pavojingos tiesiogriai sujungus baterijų **(+)** ir **(-)** terminalus. Tai yra pavojinga todėl kad elektros srovės dydis **sutrumpinimo** metu elektros srovė gali būti labai didelė, o tekančios energijos išsiskyrimas gali tapti labai "dramatiškas" (Dažniausiai šilumos pavidalu). Dažniausiai elektros grandinės yra konstruojamos taip kad ši energija yra saugiai ir praktiškai išnaudojama.


### Srovės tekmė pro Lempos filamenta

Vienas iš praktiškų ir populiariausių elektros srovės naudojimo būdu yra naudojama apšvietimui. Papraščiausias būdas norint išgauti šviesą tai yra pro elektrinę lempa kurioje yra mažas metalinis siūlas susuktas į spirale kuris yra apgaubtas permatomu stiklu. Šis filamentas švies baltai dėl karščio energijos kai pakankamai srovės pro jį tekės. Kap ir baterija ši lempa turi dvi induktyvias jungtis/terminalus vienas pro kur srovė ateis kitas pro kur ji išeis. Sujungus tokia lempa su baterija elektros grandinėje atrodys taip:

Kai Srovė teka pro lempos ploną filamentą srovė susiduria su didesniu pasipriešinimu negu įprastai susidurtu storame laide. Šis reiškinys priklauso nuo to iš kokios medžagos padarytas šis objektas, nuo diametro, ir temperatūros. Šis reiškinys vadinamas **Varža**. Galima pasakyti kad induktoriai turi labai maža varžą o izoliatoriai turi labai didelę varžą. <br> Ši varža yra naudojama norint sumažinti akumuliatoriaus (Baterijos) teikiamai įtampai.. Kai srovė susiduria su varža, susidaro "trintis". Lygiai taip pat kaip mechaninė trintis, elektros grandinėse susidariusi trintis pavirsta karščiu. Šis šilumos energija yra pakankama kad tai sukeltu ploname lempos filamente šviesą. o patys laidai kurie jungiasi su lempa net neįšils dėl savo mažos varžos. nors ir elektros srovė (Elektronų srautas) pro juos teka ta pati. 

Lygiai taip pat kai prieš tai matytame pavizdyje jei mes turėsime nutrūsią grandine **(atvirą grandine)** srovė grandinėje nebesusidarys ir lempute nebešvies. Vėl sujungus laidus kad grandinė sudarytu **uždarą grandinę** ir lempute vėl pradės šviesti

Įprastai jeigu mes norime įjungti ir išjunti šviesas savo namuose ar mūsu sukonstruotoje elektros grandinėje įprastai mes naudojam jungiklius kurie mūsu grandinę padaro uždarą arba atvirą. Šis jungiklis elektros schemoje atrodo taip:

### Įtampa ir srovė praktinėje grandinėje

Kadangi reikia energijos kad elektros srovė tekėtų pro varžą esančią visoje Grandinėje, tai grandinėje įtampa bus skirtinga tarp skirtingų grandinės atkarpų. Tarp kekvienos šios atkarpos bus varža. <br> Svarbu suprasti tai kad nors ir srovė paprastoje grandinėje bus lygiai tokia pati betkuriame grandinės taške įtampa tuo tarpu gali drastiškai keistis. 

Paimkime šita grandinę kaip pavizdį. Ant šios grandinės yra pažymėti 4 taškai (1;2;3;4) jeigu mes pamatuotume srovę tekančią laido atkarpa (1;2) srovė bus lygiai tokia pati kaip ir tekanti atkarpose (2;3), (3;4). <br> Bet mes pamatysime kad įtampa tarp betkurio šio taško bus skirtinga ir bus lygiai proporciška induktyvioje medžiagoje esančiai varžai (jeigu pati srovė šioje paprastoje grandinėje bus tokia pati, o ši grandinė tokia ir yra)

## Omo dėsnis - Kaip Įtampa Srovė ir Varža vienas kitą veikia. 

Pats pagrindinis ir tikriausiai svarbiausias dėsnis elektronikoje yra omo dėsnis kuris nurodo sąveiką tarp Srovės, Įtampos ir Varžos. Šis dėsnis yra pavadintas Omo desniu nes jį atrado **Georg Simon Ohm** ir buvo paskelbtas 1827 metais. 

### Įtampa, Srovė, Varža

Elektrinė grandinė yra suformuojama tada kai induktyvus kelias yra suformuojamas ir elektronai gali laisvai keliauti laidu. Šis nuoatinis elektronų judėjimas pro induktyvų laidą vadinamas **Įtampa**, ir dažniausiai išsireiškiama kai elektros tėkmė, lygiai taip pat kaip skystis judantis tuščiaviduriu vamzdžiu. 

Jėga kuri "stumia" elektronus kad jie "tekėtų" vadinama Įtampa ir ji yra matuojama Voltais. Įtampa matuoja potencialią energija kuri  visuomet yra matuojama tarp dviejų taškų. 

Kai kalbame apie tam tikrą įtampos kiekį grandinėje, mes kalbame apie tai, kiek potencialios energijos yra grandinėje norint "stumti" elektronus nuo vieno taško grandinėje iki kito. Jeigu minėtume "Įtampa" kalbant apie vieną tažką vadinasi kažką suprantame negerai arba įvyko kažkokia kita žmogiška klaida. 

Srovė Bandydama tekėti pro induktorių visafa susidurs su ta tikru pasipriešinimu "trintimi". Ši Trintis arba pasipriešinimas yra vadinama Varža. Srovės kiekis grandinėje yra priklausomas nuo Įtampos (potencialios energijos) kiekio ir kiek šioje grandinėje egzistuoja pasipriešinimo (Varžos). 

Lygiai taip pat kaip Įtampa, varža irgi yra matuojama tarp dviejų taškų. Dėl šios priežasties kai kalbame apie Įtampa arba stovę mes visada nurodome šias vertes sakydami "Tarp" dviejų taškų grandinėje

### Kulonas ir Elektrinis krūvis

Dar vienas iš elektronikos indžinerijoje naudojamų matavimo vienetų yra Kulonas (Ang. Coulomb). Simbolis **[q]**, žymėjimao vienetas **[C]**. Šis matmuo nurodo elektros krūvio dydi proporcingą tam tikram elektronų kiekiui. Lygiai taip pat kaip mes turime nustate kiek sveria kilogramas arba koks yra viens centimetras ilgio taip ir vienas kulono krūvis **[1C]** yra lygus **6,250,000,000,000,000,000** elektronų nešamo krūvio. tai vieno elektrono krūvis yra lygus:

$$
e = 0,00000000000000000016C
$$

$$
e = 1,6x10^{-19}C
$$

Taipogi taip gaunasi kad vienas kulonas [1C] yra lygus vienam amperui [1A] tekančiame elektrinės grandinės viename taške per sekunde:

$$
1A =\frac{6,250,000,000,000,000,00}{1sec}
$$

Kaip ir minėta prieš tai Įtampa yra potencialios energijos matmuo, Skirtas "stumti" srovės srautui iš vieno taško į kitą. Kad galėtume tiksliai apibrėžti, kas yra „voltas“, turime suprasti, kaip išmatuoti šį kiekį, kurį vadiname „potencialia energija“. Bendrasis vienetas bet kokios rūšies energijos yra džaulis *[j]*, lygus darbo kiekiui, kurį atlieka 1 Niutono jėga, judanti 1 metrą (ta pačia kryptimi). 

## Omo dėsnis

**Georg Simon Ohm** Pagrindinis atradimas buvo tai kad **elektros srovė (I)** pro metalinį laidininką grandinėje yra tiesiogiai proporcingas **Įtampai(V)**, esant betkokioje temperatūroje. Jis išreiškė savo atradimą paprasta formule:

$$
V=IR
$$

Šioje algebrinėje išraiškoje įtampa (E) yra lygi srovei (I), padaugintai iš varžos (R). Naudodami algebros metodus, galime manipuliuoti šia lygtimi į du variantus, atitinkamai išspręsdami I ir R:

$$
I=\frac{V}{R} 
$$

$$
R=\frac{E}{I}
$$
